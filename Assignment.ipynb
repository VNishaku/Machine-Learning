Q4. Write a menu driven program to reverse the entered numbers and print the sum of digits entered.

```python
while True:
    num_str = input("Enter a number (or 'q' to quit): ").strip()
    if num_str.lower() == "q":
        break

    # handle optional sign
    sign = ""
    if num_str.startswith(("+", "-")):
        sign = num_str[0]
        num_body = num_str[1:]
    else:
        num_body = num_str

    if not num_body.isdigit():
        print("Please enter a valid integer.")
        continue

    reversed_str = num_body[::-1].lstrip("0") or "0"
    reversed_num = int(sign + reversed_str)
    digit_sum = sum(int(d) for d in num_body)

    print("Reversed number:", reversed_num)
    print("Sum of digits:", digit_sum)
    print()
```

Q5.Write a menu driven program to enter the number and print whether the number is odd or even
prime.

```python
while True:
    s = input("Enter an integer (or 'q' to quit): ").strip()
    if s.lower() == "q":
        break
    if not s.lstrip("+-").isdigit():
        print("Please enter a valid integer.")
        continue

    num = int(s)
    parity = "even" if num % 2 == 0 else "odd"

    if num > 1:
        is_prime = True
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
    else:
        is_prime = False

    print(f"{num} is {parity}.")
    print(f"{num} is {'prime' if is_prime else 'not prime'}.")
    print()
```

Q8. Write a program to check if the entered number is Armstrong or not.

```python
while True:
    s = input("Enter a number (or 'q' to quit): ").strip()
    if s.lower() == "q":
        break
    if not s.lstrip("+-").isdigit():
        print("Please enter a valid integer.")
        continue

    n = int(s)
    if n < 0:
        print("Negative numbers are not considered Armstrong numbers.")
        continue

    digits = s.lstrip("+")
    power = len(digits)
    armstrong_sum = sum(int(d) ** power for d in digits)

    if armstrong_sum == n:
        print(f"{n} is an Armstrong number.")
    else:
        print(f"{n} is not an Armstrong number.")
    print()
```

Q9. Write a program to find factorial of the entered number using recursion.

```python
def factorial(n: int) -> int:
    if n <= 1:
        return 1
    return n * factorial(n - 1)

while True:
    s = input("Enter a non-negative integer to compute factorial (or 'q' to quit): ").strip()
    if s.lower() == "q":
        break
    if not s.isdigit():
        print("Please enter a valid non-negative integer.")
        continue

    n = int(s)
    print(f"{n}! =", factorial(n))
    print()
```

Q15. Write a python function sin (x, n) to calculate the value of sin(x) using its Taylor series
expansion up to n terms.

```python
import math

def sin(x: float, n: int) -> float:
    """Compute sin(x) using Taylor series expansion up to n terms."""
    if n <= 0:
        return 0.0
    total = 0.0
    sign = 1
    power = x
    fact = 1
    for k in range(n):
        total += sign * power / fact
        sign *= -1
        power *= x * x
        fact *= (2 * k + 2) * (2 * k + 3)
    return total
sin(4,5)
```

Q16. Write a program to determine EOQ using various inventory models.

```python
import math

def eoq_basic(D: float, S: float, H: float) -> float:
    """Basic EOQ: sqrt(2 * D * S / H)."""
    return math.sqrt(2 * D * S / H)

def eoq_with_shortage(D: float, S: float, H: float, b: float) -> float:
    """EOQ with allowed shortages (backordering cost b per unit per year)."""
    return math.sqrt((2 * D * S / H) * ((H + b) / b))

def epq(D: float, S: float, H: float, P: float) -> float:
    """EOQ for production model (EPQ), where P > D."""
    return math.sqrt((2 * D * S / H) * (P / (P - D)))

def prompt_float(prompt: str) -> float:
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("Please enter a valid number.")

while True:
    print("\nSelect inventory model:")
    print("1. Basic EOQ")
    print("2. EOQ with shortages (backorders)")
    print("3. Production EOQ (EPQ)")
    print("q. Quit")
    choice = input("Choice: ").strip().lower()

    if choice == "q":
        break

    if choice not in {"1", "2", "3"}:
        print("Invalid choice.")
        continue

    D = prompt_float("Annual demand (D): ")
    S = prompt_float("Ordering/setup cost per order (S): ")
    H = prompt_float("Holding cost per unit per year (H): ")

    if choice == "1":
        q = eoq_basic(D, S, H)
        print(f"EOQ (basic) = {q:.4f}")

    elif choice == "2":
        b = prompt_float("Shortage/backorder cost per unit per year (b): ")
        q = eoq_with_shortage(D, S, H, b)
        print(f"EOQ with shortages = {q:.4f}")

    elif choice == "3":
        P = prompt_float("Production rate (units per year, P > D): ")
        if P <= D:
            print("Production rate must be greater than demand.")
            continue
        q = epq(D, S, H, P)
        print(f"Production EOQ (EPQ) = {q:.4f}")
```

Q17. Write a program to determine different characteristics using various queuing models.

```python
def prompt_float(prompt: str) -> float:
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("Please enter a valid number.")

def prompt_int(prompt: str) -> int:
    while True:
        try:
            return int(input(prompt))
        except ValueError:
            print("Please enter a valid integer.")

def mm1(lmbda: float, mu: float):
    rho = lmbda / mu
    if rho >= 1:
        return None
    L = rho / (1 - rho)
    Lq = rho * rho / (1 - rho)
    W = L / lmbda
    Wq = Lq / lmbda
    return {"rho": rho, "L": L, "Lq": Lq, "W": W, "Wq": Wq}

def mmc(lmbda: float, mu: float, c: int):
    rho = lmbda / (c * mu)
    if rho >= 1:
        return None
    # P0
    sum_term = sum((c * rho) ** n / math.factorial(n) for n in range(c))
    term_c = (c * rho) ** c / (math.factorial(c) * (1 - rho))
    P0 = 1 / (sum_term + term_c)
    Lq = P0 * ( (c * rho) ** c * rho ) / (math.factorial(c) * (1 - rho) ** 2)
    L = Lq + lmbda / mu
    Wq = Lq / lmbda
    W = L / lmbda
    return {"rho": rho, "P0": P0, "L": L, "Lq": Lq, "W": W, "Wq": Wq}

def mm1k(lmbda: float, mu: float, K: int):
    rho = lmbda / mu
    if rho == 1:
        P0 = 1 / (K + 1)
        L = K / 2
    else:
        P0 = (1 - rho) / (1 - rho ** (K + 1))
        L = (rho * (1 - (K + 1) * rho ** K + K * rho ** (K + 1))) / ((1 - rho) * (1 - rho ** (K + 1)))
    Lq = L - (1 - P0)
    W = L / (lmbda * (1 - (rho ** K) * P0))
    Wq = Lq / (lmbda * (1 - (rho ** K) * P0))
    return {"rho": rho, "P0": P0, "L": L, "Lq": Lq, "W": W, "Wq": Wq}

while True:
    print("\nQueuing model calculator:")
    print("1. M/M/1")
    print("2. M/M/c")
    print("3. M/M/1/K")
    print("q. Quit")
    choice = input("Choice: ").strip().lower()

    if choice == "q":
        break

    if choice not in {"1", "2", "3"}:
        print("Invalid choice.")
        continue

    lam = prompt_float("Arrival rate (λ): ")
    mu = prompt_float("Service rate (μ): ")

    if choice == "1":
        res = mm1(lam, mu)
        if res is None:
            print("System is unstable (ρ >= 1).")
            continue
        print(f"ρ = {res['rho']:.4f}")
        print(f"L (avg # in system) = {res['L']:.4f}")
        print(f"Lq (avg # in queue) = {res['Lq']:.4f}")
        print(f"W (avg time in system) = {res['W']:.4f}")
        print(f"Wq (avg time in queue) = {res['Wq']:.4f}")

    elif choice == "2":
        c = prompt_int("Number of servers (c): ")
        if c <= 0:
            print("Number of servers must be positive.")
            continue
        res = mmc(lam, mu, c)
        if res is None:
            print("System is unstable (ρ >= 1).")
            continue
        print(f"ρ = {res['rho']:.4f}")
        print(f"P0 (empty system) = {res['P0']:.4f}")
        print(f"L (avg # in system) = {res['L']:.4f}")
        print(f"Lq (avg # in queue) = {res['Lq']:.4f}")
        print(f"W (avg time in system) = {res['W']:.4f}")
        print(f"Wq (avg time in queue) = {res['Wq']:.4f}")

    elif choice == "3":
        K = prompt_int("System capacity (K): ")
        if K <= 0:
            print("Capacity must be positive.")
            continue
        res = mm1k(lam, mu, K)
        print(f"ρ = {res['rho']:.4f}")
        print(f"P0 (empty system) = {res['P0']:.4f}")
        print(f"L (avg # in system) = {res['L']:.4f}")
        print(f"Lq (avg # in queue) = {res['Lq']:.4f}")
        print(f"W (avg time in system) = {res['W']:.4f}")
        print(f"Wq (avg time in queue) = {res['Wq']:.4f}")
```

Q18. Write a program to implement Inheritance. Create a class Employee inherit two classes
Manager and Clerk from Employee.

```python
class Employee:
    def __init__(self, emp_id: int, name: str, salary: float):
        self.emp_id = emp_id
        self.name = name
        self.salary = salary

    def display(self):
        return f"ID: {self.emp_id}, Name: {self.name}, Salary: {self.salary:.2f}"


class Manager(Employee):
    def __init__(self, emp_id: int, name: str, salary: float, department: str, bonus: float):
        super().__init__(emp_id, name, salary)
        self.department = department
        self.bonus = bonus

    def display(self):
        base = super().display()
        return f"{base}, Dept: {self.department}, Bonus: {self.bonus:.2f}"


class Clerk(Employee):
    def __init__(self, emp_id: int, name: str, salary: float, shift: str, hours_per_day: float):
        super().__init__(emp_id, name, salary)
        self.shift = shift
        self.hours_per_day = hours_per_day

    def display(self):
        base = super().display()
        return f"{base}, Shift: {self.shift}, Hours/Day: {self.hours_per_day:.1f}"


# Example usage
mgr = Manager(emp_id=101, name="Alice", salary=75000, department="Sales", bonus=5000)
clk = Clerk(emp_id=102, name="Bob", salary=32000, shift="Morning", hours_per_day=8)

print(mgr.display())
print(clk.display())
```

Q19. Write a program to fit Poisson distribution on a given data.

```python
data_str = input("Enter nonnegative integer data separated by spaces or commas: ").strip()
tokens = data_str.replace(",", " ").split()

data = []
for t in tokens:
    if t:
        try:
            v = int(t)
        except ValueError:
            print(f"Ignoring invalid token: {t}")
            continue
        if v < 0:
            print(f"Ignoring negative value: {v}")
            continue
        data.append(v)

if not data:
    print("No valid data entered.")
else:
    lam_poisson = sum(data) / len(data)
    obs = {}
    for v in data:
        obs[v] = obs.get(v, 0) + 1

    max_k = max(obs)
    total = len(data)

    print(f"Estimated λ (mean) = {lam_poisson:.4f}")
    print("k  observed  expected")
    chi2 = 0.0
    for k in range(0, max_k + 1):
        expected = total * math.exp(-lam_poisson) * lam_poisson**k / math.factorial(k)
        observed = obs.get(k, 0)
        if expected > 0:
            chi2 += (observed - expected) ** 2 / expected
        print(f"{k:2d} {observed:8d} {expected:9.3f}")

    print(f"Chi-square statistic: {chi2:.4f}")
```

Q20. Write a program to implement linear regression using python.

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset
X = np.array([1, 2, 3, 4, 5])   # independent variable
Y = np.array([2, 4, 5, 4, 5])   # dependent variable

# Step 1: Calculate means
X_mean = np.mean(X)
Y_mean = np.mean(Y)

# Step 2: Calculate slope (m) and intercept (b)
numerator = np.sum((X - X_mean) * (Y - Y_mean))
denominator = np.sum((X - X_mean) ** 2)
m = numerator / denominator
b = Y_mean - m * X_mean

print(f"Slope (m): {m}")
print(f"Intercept (b): {b}")

# Step 3: Predict values
Y_pred = m * X + b

# Step 4: Plot results
plt.scatter(X, Y, color="blue", label="Data Points")
plt.plot(X, Y_pred, color="red", label="Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

Q29. Write programs making use of PuLp library.

```python
import pulp

# Define the problem
model = pulp.LpProblem("Maximize_Profit", pulp.LpMaximize)

# Decision variables
x = pulp.LpVariable('x', lowBound=0, cat='Continuous')
y = pulp.LpVariable('y', lowBound=0, cat='Continuous')

# Objective function
model += 40*x + 30*y, "Profit"

# Constraints
model += 2*x + y <= 40   # Resource 1
model += x + 2*y <= 50   # Resource 2

# Solve
model.solve()

print("Status:", pulp.LpStatus[model.status])
print("Optimal x =", x.value())
print("Optimal y =", y.value())
print("Maximum Profit =", pulp.value(model.objective))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson

lambda_val = 3  
k = np.arange(0, 10)  
pmf = poisson.pmf(k, lambda_val)

plt.figure(figsize=(8, 6))
plt.bar(k, pmf, color='lightgreen', edgecolor='black')
plt.title('Poisson Distribution PMF (λ=3)', fontsize=14)
plt.xlabel('Number of events (k)', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

cdf = poisson.cdf(k, lambda_val)

plt.figure(figsize=(8, 6))
plt.plot(k, cdf, color='purple', marker='o', linestyle='-', linewidth=2)
plt.title('Poisson Distribution CDF (λ=3)', fontsize=14)
plt.xlabel('Number of events (k)', fontsize=12)
plt.ylabel('Cumulative Probability', fontsize=12)
plt.grid(True)
plt.show()

probability_4_events = poisson.pmf(4, lambda_val)
print(f'Probability of exactly 4 events: {probability_4_events:.4f}')
```
